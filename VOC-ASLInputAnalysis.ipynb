```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os
import numpy as np
from IPython.display import Audio, display

def plot_spectrogram(folder, id_):
    """
    Plots the spectrogram of a .wav file and displays an audio player,
    given the folder and ID.

    Args:
        folder (str): The name of the subfolder containing the audio file (e.g., "phonationA").
        id_ (str): The ID of the subject (e.g., "PZ001").
    """
    path = "/content/drive/MyDrive/SAND üó£Ô∏è‚Äã/VOC-ALS Database"
    file = f"{id_}_{folder}.wav"
    file_path = os.path.join(path, folder, file)

    try:
        # Load the audio file
        y, sr = librosa.load(file_path)

        # Compute the spectrogram
        D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

        # Plot the spectrogram
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Spectrogram of {file}')
        plt.tight_layout()
        plt.show()

        # Display audio player
        print("Audio Player:")
        display(Audio(y, rate=sr))


    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage (replace 'PZ001' with the actual string ID):
# plot_spectrogram(folder="phonationA", id_="PZ001")
```

```python
def plot_spectrogram_over_all_tests(id_):
    """
    Plots the spectrogram of a .wav file for each test and displays an audio player,
    given the ID.

    Args:
        id_ (str): The ID of the subject (e.g., "PZ001").
    """
    for folder in ["phonationA", "phonationE", "phonationI", "phonationO", "phonationU", "rhythmKA", "rhythmPA", "rhythmTA"]:
        plot_spectrogram(folder=folder, id_=id_)
```

# Analysis:



## Vowels
### Formants
- **Control Subjects**
  - Constant frequency
- **ALS Subjects**
  - Continuous and irregular frequency oscillations
  - Frequency Drift
  - Reduced definition towards the end of the vocal emission

### Prosody
- **Control Subjects**
  - Absence of pauses
- **ALS Subjects**
  - Occasional interruptions in pronunciation

---

## Syllables

- **Control Subjects**
  - (Not analyzed, but standard pronunciation is assumed)
- **ALS Subjects**
  - Complete lack of pulsives, or substitution with a nasal consonant (e.g. /k/ > /m/, /t/ > /n/)
  - Lenition/distortion of pulsives into approximants
  - Palatalization/alveolarization and affrication of velar pulsives (e.g. /k/ > /dr/ or /d í/)
  - Voicing of velar pulsives (e.g. /k/ > /g/)
  - Pronunciation of alveolar (/t/) or bilabial (/p/) pulsives may be completely unaffected
