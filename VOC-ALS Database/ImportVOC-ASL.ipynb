```python
 import synapseclient
 import synapseutils

 syn = synapseclient.Synapse()
 syn.login(authToken=MASTER_TOKEN)
```

```python

def copy_file_to_destination(file_to_save, subfolder):
  # Assuming 'files' is the list of downloaded Synapse files and the first element is the one you want
  # The file path in the cache is available in file_info.path
  source_path = file_to_save.path
  filename = file_to_save.name # Get the original filename with extension

  # Construct the destination path in Google Drive
  drive_subfolder = f"{drive_folder}/{subfolder}"
  os.makedirs(drive_subfolder, exist_ok=True)
  destination_path = os.path.join(drive_subfolder, filename)

  # Copy the file from the Synapse cache to Google Drive
  shutil.copy(source_path, destination_path)

  print(f"File '{filename}' copied to: {destination_path}")
```

```python
for subfolder, id in FILE_MAP.items():
  files = synapseutils.syncFromSynapse(syn, id)
  for file in files:
    copy_file_to_destination(file, subfolder)
```
